```python
import requests
from bs4 import BeautifulSoup
import html5lib
from pprint import pprint

url = "https://www.giveindia.org/certified-indian-ngos"

def environment(nameList):
    name = ""
    index = 0
    while(index < (len(nameList))):
        if(nameList[index] == "|"):
            break
        else:
            name = name + nameList[index]
            name = name + " "
        index = index + 1
    return name

def state(nameList):
    name = ""
    index = 0
    while index<len(nameList):
        if(nameList[index]=="|"):
            num = index+1
            break
        index = index + 1
    while num < len(nameList):
        name = name + nameList[num]
        name = name + " "
        num = num + 1

    return name

def DetailsOfNgo(url):
    data = requests.get(url).text
    soup = BeautifulSoup(data, 'html5lib')
    ngos_class = soup.findAll('div', {'class':'nonprofit-card-container d-flex py-2 container'})
    detailsList = []
    for index in ngos_class:
        allDetail = {}
        allNgoData = index.find('div', {'class':'d-flex f-d-col col-10 col-sm-10'})
        ngoName = allNgoData.find('h5', {'class':'nonprofit-name mb-0'})
        ngo = ngoName.get_text()

        div = allNgoData.find('div')
        span = div.find('span')
        name_list = span.get_text()
        name_list = name_list.split()
        cause = environment(name_list)
        
        sname = state(name_list)
               
        allDetail["NGO Name"] = ngo
        allDetail["For cause"] = cause
        allDetail["State/Province"] = sname

        detailsList.append(allDetail)
    return detailsList

details = DetailsOfNgo(url)
pprint (details)
```
